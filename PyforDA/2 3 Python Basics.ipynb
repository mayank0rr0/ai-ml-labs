```python
# Generator 
# to construct new iterable objects 
# it gives multiple values in a sequence by pausing and resuming each time called
# generator functions have yeild keyword instead of return

def gen(x):
    for i in range(1, x+1): # loop is used and inside it we have yield
        yield i**2

# calling generator function does not yeild results
gen(5)

# they must be materialized or used for requesting objects as iterable
```

        - itertool module

```python
# module has a collection of generators for common data algorithms
import itertools
```

```python
# groupby - takes any sequence and a function and group consecutive elements according to the return value of the function
names = ["Alan", "Aman", "Supriya", "Surya", "Sanjay", "Aanand", "Kumar", "Faltu", "Fahid"]

for a,y in itertools.groupby(names, lambda x: x[0]):
    print(a, list(y))
```

```python
[ (a, list(b)) for a, b in itertools.groupby(names, lambda x: x[0])]
```

        - useful itertool functions

```python
# chain(*iterambles) - gives a sequence by chaining iterables together
list(itertools.chain(names, collection))
```

```python
# combinations(iterable, k) - gives a sequence of all possible k - tuples of elements in the iterable - no order and no replacement
list(itertools.combinations(collection, 2))
# for relpacement use - combinations_with_replacement
```

```python
# permutations(iterable, k) - gives a sequence of all possible k - tuples of elements in the iterable - with order and no replacement
list(itertools.permutations(collection, 2))
```

```python
# product(*iterables, repeat=1) - generates cartesian product of input iterables as tuples, similar to nested for loop

list(itertools.product(names, collection, repeat=1))
```
