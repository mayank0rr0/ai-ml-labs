### **BASIC INDEXING** and **SLICING**

- Many ways for indexing [selecting an element or a group] - so deep topic

````python
# for one dimensional arrays ( simple and similar to Python lists )
arr = np.arange(20)
arr
````

````python
int(arr[5]) # taking out an element and convering to int data type
````

````python
arr[5:9] # slicing a part of the array
````

````python
arr[2:8] = 12 # assigning a part of the array to specific number - each element in the slice will be equal to that
# the value was broadcasted/ prpagated across the entire selection or slice
arr
````

Here the slice is a view/ image of the original array
 - data in not copied 
 - any modification to the view will be reflected

````python
# bare slice [:] - assigns to all values of an array
arr[:] = 5 
arr
````

As numPy works with very large arrays thus creating copies would affect performance and memory problems 
  - hence the view creation
  - copying should be performed using .copy() method on the slice
  

````python
# for higher dimensional arrays - they  need to be accessed recursively
arr2 = np.array([[1,2,3],[1,5,8]])
````

````python
arr2[0][2] # defining a path to it form out to in
````

````python
arr2[0,2] # defifnign a path is a comma separated value
````

````python
arr2[0] # if lower/inner indices are ommited then a lower dimension ndarray is given rather than specific element
````

````python
# assigning to the selected element
old_val = arr2[0].copy()
# scalar can be assigned to brodcast the value to all underlying elements
arr2[0] = 42
arr2 
````

````python
# arrays can be assigned to replace the underlying elements
arr2[0] = old_val   
arr2
````

````python
# Boolean conditions can be combinied by (and - & ) ; (or - |)
arr2[(arr1 == "bob") | (arr1 == "puss")]
````

- Indexing using integer arrays

````python
arr[[2,3,6], [1,3,2]] # first array traverses row and the next traverses columns
````

````python
arr[[2,3,6]][:, [1,3,2]] 
# this first extrats the row as per the first array then it extrats all columns for those rows
# after that the second array is used to select and order the extracted row and a rectangular region is formed 
````

### **UNIVERSAL FUNCTIONS** - Element-Wise Array Functions

```ufunc``` - function that performs element-wise operations on data in ndarrays
- fast vectorized wrappers for simple functions that take one or more scalar values and produce one or more scalar results
- simple element-wise transformations eg: ```numpy.sqrt``` ; ```numpy.exp```
- unary ufuncs - take single input
   - ```.sqrt```
   - ```.exp```

- binary ufuncs - take multiple output 
   - ```.maximum()```
   - ```.minimum()```
   - ```.add()```
   
- function that returns multiple arrays - ```np.modf(arr)``` -> gives a fractional part and integral part floating point arrays\

- ufunc has option ```out``` argument to direct output to an existing array rather than a new one

- New ufuncs keep coming so look out @ numpy website

### **ARRAY-ORIENTED** Progamming

numpy arrays- can perform data processing tasks as concisily that might require a for loop
- replacing explicit lopps with array expression as vectorization
- broadcasting can be used for vectorizations

```mesh-grid``` - function takes one or two arrays and produces two 2D matrices

**EXAMPLE**

Evaluate the function sqrt(x^2 + y^2) across a regular grid values 

````python
points = np.arange(-2, 2, 0.01)

xs, ys = np.meshgrid(points, points)

ys
````

````python
xs
````

````python
z = np.sqrt(xs ** 2 + ys ** 2)

z
````

Expressing **CONDITIONAL LOGIC** as Array Operations

```numpy.where(CONDITION, TRUE:X , FALSE:Y)``` function is a vectorized versionn of ternary expression ```X if CONDITION else Y```

````python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, False, True])
````

````python
np.where(cond, xarr, yarr) # second and third argument can be a scalar
````

- |Used to change array according to some condition

````python
arr = rng.standard_normal((4,4)) 
arr
````

````python
arr > 0 # Condition Output
````

````python
np.where( arr<0, 0, arr ) # making all negative values as zero
````

````python
np.where( arr<0, -1, 1 ) # making all negative values -1 and possitive values 1
````
