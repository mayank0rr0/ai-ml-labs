<center>Code for Permutation</center>

```python
n = int(input("Total number of objects"))
r = int(input("Number of objects selected"))

print("Total number is " + str(n) + ".\n" + "Number selected is " + str(r) )
```

```python
# Calculating possible arrangements

# Preseving previously created variable - otherwise variable will be needed on every run 
m=n
arrangements = 1

# Iterating r times and multiplying m to arrangement while decreasing it
for i in range(r):
    # Multipying m as it is the number of options
    arrangements = m * arrangements
    # As no replacement occurs the number of options decreases 
    m = m - 1

print("Number of arrangements are " + str(arrangements))
```

<center>Code for Combination</center>

```python
# Calculating possible combinations

# Preserving variable
m = n
combinations = 1

# Iterating r times and multiplyiing m to combinations and dividing by 1 while decreasing it
for i in range(r):
    # Multipying m as it is the number of options
    # Dividing by i+1 to remove the extra counts
    combinations = ( combinations * m ) / (i + 1)
    # Decreasing m by 1
    m = m - 1

print("Number of combinations are " + str(int(combinations)))
```
