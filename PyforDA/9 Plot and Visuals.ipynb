### Figures and Subplots

**FIGURE** - object that holds the plot in matplotlib
- to create a new figure - `plt.figure`

```python
fig = plt.figure()
```

- to fix the size and aspect ratio of the figure 
- use figsize

**SUBPLOT** - plots contained in a figure

```python
# these all should be in a single cell as plots are reset when cell is reevaluated 
fig = plt.figure()

# in (2,2,1) first part defines the size 2x2 and the last one gives the cardinality of the subplot
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
```

- `fig.add_subplot` returns `AxesSubplot` objects

- plot axis objects hace various methods for diffeerent types of plots
   - axis method should be used over top-level plotting functions like plt.plot

```python
ax3.plot(np.random.standard_normal(50).cumsum(), color="green", linestyle="dashed")
```

```python
fig
```

- `AxesObjects` can be plotted upon by using their methods

```python
ax1.hist(np.random.standard_normal(100), bins=20, color="blue", alpha=0.3) # alpha=0.3 means the opacity of the plot
```

```python
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.standard_normal(30))
```

```python
ax4.hist(np.random.standard_normal(40), bins=4, color="black")
```

```python
fig
```

Adjusting the Spacing and Subplots

- Default spacing can be changes using the `subplots_adjust` methods on Figure object

`subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)`

- wspace and hspace control the percent of the figure width and figure height

```python
fig, axes = plt.subplots(2,2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.standard_normal(500), bins=50, color="black", alpha=0.5)

fig.subplots_adjust(wspace=0, hspace=0)
```

- here the tick locations and labels are needed to be fixed
   - matplotlib doesnot check overlapping labels

### **Ticks, Legends** and **Labels**

plot decoration in matplotlib thru - methods
 - `xlim` - plot range
 - `xticks` - tick location
 - `xlabels` - tick labels

- called with no argument they return the current value
   - `ax.xlim()` - x axis plotting range -- corresponds to `ax.get_xlim`
- called with parameter value sets the range
   - `ax.xlim([ 0,10 ])` - sets range as 0 to 10 -- corresponds to `ax.set_xlim`
- all such methods act on active or most recent AxesSubplot

#### Setting title, axis labels, ticks and tick labels

```python
fig, ax = plt.subplots()

ax.plot(np.random.standard_normal(1000).cumsum());
```

```python
# `set_xticks` defines the place to put the ticks along the data range
# by default xticks are the loction of labels
# but other values can be defined using the `set_xticklabels`

ticks = ax.set_xticks([0, 250, 500, 750, 1000])

labels = ax.set_xticklabels(["one", "two", "three", "four", "five"], rotation=30, fontsize=8) 

# rotation sets x - tick labels at 30 degree rotation

ax.set_xlabel("Stages") # name to the x-axis
ax.set_title("My first matplot plot with title") # to set title of the plot
```

```python
fig
```

- Y AXIS can be modified in the same way by replacing X with Y 
- using `set` method of the axes class 
   - batch setting of plot properties can be performed

```python
ax.set(title="First matplot second title", xlabel="Stages again")
```

```python
fig
```

- `set_xlim` and `set_ylim` methods to manually set the start and end boundaries for the plot rather than using matplotlib's default
- `ax.set_title` adds a main title tot the plot
- matplotlib has objects called `patches`
   - some like `rectangle` and `circle` are in `matplotlib.pyplot`
   - others in `matplotlib.patches`

```python
fig, ax = plt.subplots()

rect = plt.Rectangle((0.2,0.75), 0.4, 0.15, color="black", alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color="blue", alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35,0.4], [0.2, 0.6]], color="green", alpha=0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```

- Many plot types are assembled with patches

### Line Plots

Series and DataFrame - have `plot` attribute for making some basic plot types
   - by default it makes - line plots

For Series - 

```python
import pandas as pd
s = pd.Series(np.random.standard_normal(10).cumsum(), index=np.arange(0,100,10))
```

```python
s.plot(figsize=(5,5));
# options
# label
# ax - matplotlib subplot object to plot on; if nothing passed uses active matplotlib subplot - increases flexibility in a grid layout
# style
# kind - "area", "bar", "barh", "density", "hist", "kde", "line" [default] , or "pie"
# figsize 
# logx - True for log scaling
# logy
# use_index=False to disable passing of index to plotting
# limits can be adjusted using `xticks` and `xlim` options - same for y axis resp.
# rot - rotate tick labels
# grid - Display axis grid (off by default)
```

### Bar Plots

- `plot.bar()` and `plot.barh()` - for vertical and horizontal bar plots
- Series and DataFrame index will be used as the x (bar) and y (barh) ticks

```python
fig, axes = plt.subplots(2,1)
data = pd.Series(np.random.uniform(size=16), index=list('abcdefghijklmnop'))

data.plot.bar(ax=axes[0], color="black", alpha=0.7)
data.plot.barh(ax=axes[1], color="black", alpha=0.8)
```

### Histograms and Density Plots

- `histogram` 
- `density plot` 
   - formed by computing an estimate of a `continuous probability distribution` that might have generated the observed data
        - usually by approximating the distribution as a mixture of kernels
        - `kernels` - simpler distribution like normal distribution
        - KDE - kernel density estimate plot
        
- `series.plot.density` - 
    - requires `scipy`
- `seaborn.histplot` - for both density plot and histogram

```python
comp1 = np.random.standard_normal(200)
comp2 = 10 + 2 * np.random.standard_normal(200)

values = pd.Series(np.concatenate([comp1, comp2]))
```

```python
sns.histplot(values, bins=100, color="black")
```
