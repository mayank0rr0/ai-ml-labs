```python
import numpy as np
```

* Using Random Number Generator of NumPy to create an array of nsteps with values 1 and 0 using integers method
* Then using the where method to change 0 to -1 and others to 1
* making a cummulative array from given array usin `.cumsum()` method 
* plotting the graph using `matplotlib.pyplot()`

```python
nsteps = 1000
```

```python
rng = np.random.default_rng(seed=13245)
```

```python
draws = rng.integers(0,2, size=nsteps)
```

```python
steps = np.where(draws == 0, 1, -1)
steps
```

```python
walk = steps.cumsum() # cumulative sum of each
walk
```

```python
walk.min()
```

```python
walk.max()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(walk)
```
