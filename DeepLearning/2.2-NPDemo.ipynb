```python
import numpy as np

a = np.array([1,2,3,4])
print(a)
```

## Vectorization vs Non Vectorization

```python
import time 

a = np.random.rand(1000000)
b = np.random.rand(1000000)

tic = time.time()
c = np.dot(a,b)
toc = time.time()

print(c)
print("Vectorized verison: " + str(1000*(toc-tic)) + "ms")

# NON Vectorized version

c = 0
tic = time.time()
for i in range(1000000):
    c += a[i] * b[i]

toc = time.time()
print(c)
print("Non-Vectorized version: " + str(1000 *(toc-tic)) + "ms")
```

## Broadcasting

```python
A = np.array([
    [56.0, 0.0, 4.4, 68.0],
    [1.2, 104.0, 52.0, 8.0],
    [1.8, 135.0, 99.0, 0.9]
])

print(A)
```

```python
cal = A.sum(axis=0) # axis = 0 for vertical sum / column wise  OR 1 for row wise
print(cal)
```

```python
percentage = 100*A/cal.reshape(1,4) # whats the need for .reshape(1,4) ? redundant - but it is broadcasting cal to A
# how (3,4) be divided by (1,4) ? 
print(percentage)
```
